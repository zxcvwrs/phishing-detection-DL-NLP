3. Removing nonalphanumeric characters, special characters, punctuation marks, whitespaces caused by removal, stopwords, HTML elements.
4. Replacing hyperlinks with fixed strings
5. Tokenization based on white spaces

```python
from bs4 import BeautifulSoup
def _remove_html_tags(body):
    try:
        soup = BeautifulSoup(body, 'html.parser')
        return soup.get_text()
    except Exception as e:
        print(f'Exception occured at: {e} with {body}')
        return "to_manual_extraction"
```

```python
test1 = '<<sc<script>script>alert(1&lt)<</sc</script>/script>'
print(_remove_html_tags(test1))
```

```python
from nltk.corpus import stopwords
def _tokenize(body, stopwords=stopwords):
    body = body.split(" ")
    body = [token for token in body if not token in stopwords.words('english')]
    return body
```

```python
len(stopwords.words('english'))
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
def lemmatize_words(words):
    lemmatizer = WordNetLemmatizer()
    new_words = []
    for word in words:
        word = lemmatizer.lemmatize(word)
        new_words.append(word)
    return new_words
```

```python
import nltk
english_words = set(nltk.corpus.words.words())
def preserve_english_words(text, english_words=english_words):
    words = [word for word in text if word in english_words and len(word) != 1]
    return words
```

```python
import re
import cleantext

def prepare_body_mail(body, tokenize_flag=True):
    body = body.lower()
    body = _remove_html_tags(body)
    try:
        body = cleantext.replace_urls(body, replace_with="fixedstringurl")
        body = re.sub(r'\S+@\S+', 'fixedstringemails', body)
        body = re.sub(r"[^a-zA-Z0-9\s]", '', body) 
        body = " ".join(body.split())
    except Exception as e:
        print(f'Exception occured at: {e} with {body}')
        return 'to_manual_extraction'
    if tokenize_flag:
        body = _tokenize(body)
    return body    
```

### Jose Nazario's phishing dataset


#### 0. Load dataset and extract body

```python
jose_file_path= 'comparision_of_nlp_and_ml_methods_datasets\jose_phishing_dataset'
```

```python
import email
import pandas as pd
from collections import defaultdict
import os

res = defaultdict(list)

for file in os.listdir(jose_file_path):
    file_path = jose_file_path + '\\' + file

    with open(file_path, 'rb') as f:
        text = f.read()

    messages = text.split(b'From ')

    df_jose_unprocessed = pd.DataFrame(columns=['body'])

    for message in messages:
        msg = email.message_from_bytes(message)
        body = None
        if msg.is_multipart(): # if e-mail contains multipart
            for part in msg.walk():
                if part.get_content_type() == 'text/plain':
                    body = part.get_payload(decode=True)
        else:
            body = msg.get_payload(decode=True) # case if single part message, extract payload
        if body:
            body = body.decode(errors='ignore') # ignore if decoding errors
            lines = body.split('\n')
            for i, line in enumerate(lines):
                if not line:
                    body = '\n'.join(lines[i+1:])
                    break
            if body:
                #df_jose_unprocessed = df_jose_unprocessed.append({'body': body}, ignore_index=True)
                res['body'].append(body)

df_jose_unprocessed = pd.DataFrame(res)
```

```python
strings_to_remove = ('Content-Type', 'Content-Transfer-Encoding', 'MIME-Version', 'Subject', 'From', 'To', 'Date', 'Message-ID', 'X-Mailer', 
    'X-MimeOLE', 'X-Priority', 'X-MSMail-Priority', 'X-Originating-IP', 'X-Spam', 'X-Spam-Status',
    'text/plain', 'You will not see this in a MIME-aware mail reader.', 'jose@monkey.org', '--===============',
    "This is a multi-part message in MIME format.", 'Return-Path', 'charset=', 'Delivered-', '--', 'Content-Description: Mail message body =',
    'X-FDA:', 'Received:', 'Content-type:', 'Content-Description:', 'This is a MIME message. If you are reading this text, you may want to consider changing to a mail reader or gateway that understands how to properly handle MIME multipart messages.'
    '------=_', 'This is a MIME encoded message', 'This is a multipart message in MIME format.', 'This message is in MIME format.', 'Content-Disposition:',
    'format=', 'boundary=', 'X-HE-Tag: ', 'X-Filterd-Recvd-Size: ', 'by imf02.b.hostedemail.com', '(Postfix)', 'jose@login.monkey.org', 'monkey.org', 
    'X-Forwarded-For: Authentication-Results: ', 'Message-Id: ', 'X-Probable-Spam: ', 'X-Scanned-By: MIMEDefang 2', 'X-Scanned-By', 'X-Keywords:', 
    'X-UID: ', 'DKIM-Signature:', 'c=relaxed/relaxed')

def remove_or_extract(mail):
    for string_to_remove in strings_to_remove:
        while string_to_remove in mail:
            index = mail.find(string_to_remove)
            end_index = mail.find('\n', index)
            if end_index == -1: # if not found \n it searches for \r
                end_index = mail.find('\r', index)
            if end_index != -1:
                mail = mail[:index] + mail[end_index:]
            else:
                mail = mail.replace(string_to_remove, "")
                break
    return mail

#print(remove_or_extract(df_jose_unprocessed['body'][0]))
```
